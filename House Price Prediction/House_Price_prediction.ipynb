```python
#Feature Engineering
```

```python
df = df.drop("id",axis=1)
```

```python
df["date"] = pd.to_datetime(df['date'])
```

```python
df['year'] = df['date'].apply(lambda date: date.year)
df['month'] = df['date'].apply(lambda date: date.month)
```

```python
sns.boxplot(x='month',y='price',data=df)
```

```python
df.groupby('month').mean()['price'].plot()
```

```python
df.drop("date",axis=1,inplace=True)
```

```python
df.drop("zipcode",axis=1,inplace=True)
```

```python
df['yr_renovated'].value_counts()
```

```python
df['sqft_basement'].value_counts()
```

```python
##Train Test
x= df.drop("price",axis=1).values
y=df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
x_train=scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from keras.layers import Dense
from keras.models import Sequential
```

```python
x_train.shape
```

```python
model = Sequential()
```

```python
model.add(Dense(20,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),batch_size=128,epochs=400,verbose=1)
```

```python
df_val = pd.DataFrame(model.history.history)
df_val.head()
```

```python
df_val.plot()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,explained_variance_score
```

```python
predict = model.predict(x_test)
```

```python
mean_absolute_error(y_test,predict)
```

```python
mean_squared_error(y_test,predict)
```

```python
mean_squared_error(y_test,predict)**.5
```

```python
explained_variance_score(y_test,predict)
```
