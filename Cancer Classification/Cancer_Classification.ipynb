```python
x = df.drop("benign_0__mal_1",axis=1).values
y = df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar = MinMaxScaler()
x_train = scalar.fit_transform(X_train)
x_test = scalar.transform(X_test)
```

```python
from keras.layers import Dense,Dropout
from keras.models import Sequential
```

```python
model = Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss = 'binary_crossentropy',optimizer='adam')
```

```python
from keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
```

```python
model.fit(x_train,y_train,epochs=400,validation_data=(x_test,y_test),callbacks=[early_stop])
```

```python
losses=pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
model1 = Sequential()
model1.add(Dense(30,activation='relu'))
model1.add(Dropout(.5))
model1.add(Dense(15,activation='relu'))
model1.add(Dropout(.5))
model1.add(Dense(1,activation='sigmoid'))
model1.compile(loss = 'binary_crossentropy',optimizer='adam')
```

```python
model1.fit(x_train,y_train,epochs=600,validation_data=(x_test,y_test),callbacks=[early_stop])
```

```python
losss=pd.DataFrame(model1.history.history)
losss.plot()
```

```python
predict = model.predict_classes(x_test)
predict1 =model1.predict_classes(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predict))
print(classification_report(y_test,predict1))
```

```python
print(confusion_matrix(y_test,predict))
print(confusion_matrix(y_test,predict))
```
